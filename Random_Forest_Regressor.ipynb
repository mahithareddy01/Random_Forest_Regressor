```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.arange(1, 21).reshape(-1, 1)
y = np.array([
10, 12, 15, 18, 20, 25, 30, 28, 35, 40,
45, 42, 50, 55, 60, 58, 65, 70, 72, 75])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(
    n_estimators=100,
    random_state = 42
)
rf.fit(x_train,y_train)
```

### **Random Forest Metrics**

```python
y_pred_rf = rf.predict(x_test)
print(f"MSE: {mean_squared_error(y_test,y_pred_rf)}")
print(f"R2 score: {r2_score(y_test,y_pred_rf)}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(x_train,y_train)
```

```python
y_pred = dt_reg.predict(x_test)
```

### **Decision Tree Metrics**

```python
from sklearn.metrics import mean_squared_error,r2_score
print(f"mse: {mean_squared_error(y_test,y_pred)}")
print(f"r2 score: {r2_score(y_test,y_pred)}")
```

```python
import matplotlib.pyplot as plt
X_grid  = np.linspace(1,20,200).reshape(-1,1)
plt.scatter(X,y,color='black')
plt.plot(X_grid,dt_reg.predict(X_grid),color='red',label='Decision tree Regressor')
plt.plot(X_grid,rf.predict(X_grid),color='green',label='Random Forest Regressor')
plt.legend()
plt.show()
```
